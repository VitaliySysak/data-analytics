Написати додаток, який отримує вибірку впорядкованих пар. Кожна пара розміщена на новій строці. Перевірити чи існує кореляція між значеннями пар. Та зайти графік лінійної регресії для вказаної перевірки. Побудувати коефіцієнт детермінації, для графіка регресії, що знайдено.

```python
import pandas as pd
import math

df = pd.read_csv('./data/salaries_by_college_major.csv')
df = df[['Starting Median Salary', 'Mid-Career Median Salary', 'Group']].dropna()
```

```python
k = 3
starting_salary = 50000
mid_career_salary = 90000

values_to_classify = [starting_salary, mid_career_salary]
```

```python
dataset = []
for index, row in df.iterrows():
    features = [row['Starting Median Salary'], row['Mid-Career Median Salary']]
    label = row['Group']
    dataset.append([features, label])

distances = []

for data_point in dataset:
    training_features = data_point[0]
    training_group = data_point[1]
    dist_sq = (training_features[0] - values_to_classify[0])**2 + (training_features[1] - values_to_classify[1])**2
    distance = math.sqrt(dist_sq)
    distances.append( (distance, training_group) )

distances.sort(key=lambda x: x[0])
neighbors = distances[0:k]

votes = {} 
for (dist, label) in neighbors:
    if label in votes:
        votes[label] += 1
    else:
        votes[label] = 1

winner = max(votes, key=votes.get)

print(f"Початкова зарплата = {starting_salary}")
print(f"Зарплата в середині кар'єри = {mid_career_salary}")
print(f"{k} найближчих сусідів:")
for distance, label in neighbors:
    print(f"Клас: {label} (Відстань: {distance:.2f})")
```

```python
print(f"голоси: {votes}")
print(f"\nПрогноз: Елемент належить до групи '{winner}'.")
```
