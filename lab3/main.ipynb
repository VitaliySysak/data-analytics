Написати програму, яка вирахує довірчий інтервал для математичного спобівання і дисперсії.
Рівень значущості вказується зовні. Допустимими варіантами рівня значущості:
0.05
0.01
0.1

```python
import math
import pandas as pd
from scipy import stats
```

```python
data = pd.read_csv("./data/ScreenTime vs MentalWellness.csv")
values = data["stress_level_0_10"]
```

```python
alphas = [0.05, 0.01, 0.1]
n = len(values)
print(f"Кількість спостережень: {n}")
```

```python
for alpha in alphas:
    mean = values.mean() 
    s = values.std(ddof=1)
    t_critical = stats.t.ppf(1 - alpha/2, n - 1)
    margin = t_critical * s / math.sqrt(n)
    lower_mean = mean - margin
    upper_mean = mean + margin

    variance = values.var(ddof=1)

    left_limit = stats.chi2.ppf(alpha / 2, n - 1)
    right_limit = stats.chi2.ppf(1 - alpha / 2, n - 1)

    lower_variance = (n - 1) * variance / right_limit
    upper_variance = (n - 1) * variance / left_limit

    print(f"Рівень значущості: {alpha}")
    print(f"Вибіркове середнє = {mean:.4f}")
    print(f"{100*(1-alpha):.0f}% довірчий інтервал для середнього: ({lower_mean:.4f}, {upper_mean:.4f})")
    print()
    print(f"Вибіркова дисперсія = {variance:.4f}")
    print(f"{100*(1-alpha):.0f}% довірчий інтервал для дисперсії: ({lower_variance:.4f}, {upper_variance:.4f})")
```
