Написати додаток, який отримує дві вибірки, для цих вибірок перевірити гіпотезу, чи рівна в них дисперсія (за альтернативну гіпотезу взяти гіпотезу, що дисперсії недорівнюють одна одній). У випадку, якщо гіпотеза підтверджена перевірити гіпотезу, чи рівне в них математичне сподівання (за альтернативну гіпотезу взяти гіпотезу, що математичне сподівання недорівнють одне одному)

```python
import numpy as np
from scipy.stats import f, ttest_ind
```

```python
alpha = 0.05

sample1 = [14, 11, 12, 14, 14, 8, 12, 10, 10, 11, 10, 13, 12, 10, 11, 11, 13, 10, 11, 8, 5, 11, 12, 9, 15, 7, 10, 10, 13, 13]
sample2 = [10, 11, 8, 6, 9, 10, 12, 12, 9, 9, 8, 7, 7, 14, 9, 9, 7, 12, 7, 10, 8, 11, 9, 8, 10, 11, 10, 11, 9, 9]
```

H0 (нульова гіпотеза): дисперсії рівні

H1 (альтернативна гіпотеза): дисперсії не рівні

```python
varience1 = np.var(sample1, ddof=1)
varience2 = np.var(sample2, ddof=1)

if varience1 > varience2:
    f_statistic = varience1 / varience2
    freedom_degree1 = len(sample1) - 1
    freedom_degree2 = len(sample2) - 1
else:
    f_statistic = varience2 / varience1
    freedom_degree1 = len(sample2) - 1
    freedom_degree2 = len(sample1) - 1

p_value = f.sf(f_statistic, freedom_degree1, freedom_degree2) * 2

print(f"Вибіркова дисперсія першої вибірки: {varience1:.4f}")
print(f"Вибіркова дисперсія другої вибірки: {varience2:.4f}")
print(f"f-статистичне: {f_statistic:.4f}")
print(f"p-значення: {p_value:.4f}")

if p_value > alpha:
    print(f"({p_value:.4f}) > alpha ({alpha}). Не відхиляємо нульову гіпотезу. Вважатимемо що дисперсії рівні.")
    variances_are_equal = True
else:
    print(f"({p_value:.4f}) <= alpha ({alpha}). Не відхиляємо нульову гіпотезу. Дисперсії відрізняються.")
    variances_are_equal = False
```

```python
if variances_are_equal:
    print("Тест Стьюдента")
    t_statistic, p_value_t = ttest_ind(sample1, sample2, equal_var=True)

    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    
    print(f"Середнє значення першої вибірки: {mean1:.4f}")
    print(f"Середнє значення другої вибірки: {mean2:.4f}")
    print(f"t-статистичне: {t_statistic:.4f}")
    print(f"p-значення: {p_value_t:.4f}")

    if p_value_t > alpha:
        print(f"p-значення ({p_value_t:.4f}) > alpha ({alpha}).")
        print("Немає підстав відхилити нульову гіпотезу. Вважаємо, що математичні сподівання рівні")
    else:
        print(f"({p_value_t:.4f}) <= alpha ({alpha}). Відхиляємо нульову гіпотезу. Математичні сподівання не рівні")
else:
    print("Дисперсія не підтверджена")
```
