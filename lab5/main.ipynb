Написати додаток, який отримує вибірку впорядкованих пар. Кожна пара розміщена на новій строці. Перевірити чи існує кореляція між значеннями пар. Та зайти графік лінійної регресії для вказаної перевірки. Побудувати коефіцієнт детермінації, для графіка регресії, що знайдено.

```python
import pandas as pd
import math

df = pd.read_csv('./data/salaries_by_college_major.csv')
df = df[['Starting Median Salary', 'Mid-Career Median Salary']].dropna()
```

```python
start_median_salary = df['Starting Median Salary']
mid_career_median_salary = df['Mid-Career Median Salary']

n = len(start_median_salary)

start_median_salary_mean = sum(start_median_salary) / n
mid_career_median_salary_mean = sum(mid_career_median_salary) / n

# r = Σ((x - x̄)(y - ȳ)) / sqrt(Σ(x - x̄)^2 * Σ(y - ȳ)^2)
r = ((start_median_salary - start_median_salary_mean) * (mid_career_median_salary - mid_career_median_salary_mean)).sum() / math.sqrt(((start_median_salary - start_median_salary_mean)**2).sum() * ((mid_career_median_salary - mid_career_median_salary_mean)**2).sum())

print(f"Коефіцієнт кореляції r: {r:.4f}")
```

```python
# b1 = Σ((x - x̄)(y - ȳ)) / Σ((x - x̄)^2)
# b0 = ȳ - b1*x̄
b1 = ((start_median_salary - start_median_salary_mean) * (mid_career_median_salary - mid_career_median_salary_mean)).sum() / ((start_median_salary - start_median_salary_mean)**2).sum()
b0 = mid_career_median_salary_mean - b1 * start_median_salary_mean

print(f"Рівняння регресії: y = {b0:.2f} + {b1:.2f}x")
```

```python
# R² = r²
R2 = r**2
print(f"Коефіцієнт детермінації (R²): {R2:.4f}")
```
