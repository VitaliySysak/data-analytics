Додаток отримує на вхід csv файл зі значеннями експерименту. Вважати, що значення є дискретними.
Дивлячись на фактичні данні експерименту програма виводить математичне сподівання, дисперсію
та середньоквадратичне відхилення вибірки, що присутня у файлі

```python
file_path = "./data/ScreenTime vs MentalWellness.csv"
df = pd.read_csv(file_path)
stress_level = df["stress_level_0_10"]
```

```python
counts = df["stress_level_0_10"].value_counts()
n = len(df["stress_level_0_10"])
mean_manual = sum(x * (count / n) for x, count in counts.items())
```

```python
print(f"Математичне сподівання: {mean_manual}")
```

```python
variance = sum((x - mean_manual) ** 2 for x in stress_level) / n
```

```python
print(f"Дисперсія: {variance}")
```

```python
std_sample = variance ** 0.5
```

```python
print(f"Середньоквадратичне відхилення: {std_sample}")
```
